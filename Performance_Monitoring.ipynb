The first Component we are going to Analyze is the **Processor**. The **Processor**, more commonly known as **CPU (Central Processing Unit)** is the logical and physical unit or subsystem that oversees the functionalities of a computer .

The **CPU** has the task of **Executing the Instructions** of a program present in the **ROM** during boot, in the central memory (or primary, e.g. **RAM** ).

The **CPU** is based on the following cycle:
- **Instruction Fetch:** The Processor fetches the instruction from the memory, present in the address (typically logical) specified by a "specific" register.
- **Decoding:** Once the word has been picked up, it is determined which operation must be performed and how to obtain the operands.
- **Execute:** The desired computation is performed. In the last step of the execution the PC is increased : typically by one if the instruction was not a conditional jump, otherwise the increment depends on the instruction and on the outcome of this.

## 1.1.2: K-MEANS CLUSTERING & ELBOW METHOD:

##### INTRODUCTION:

The **K-MEANS CLUSTERING** Algorithm is an **Unsupervised Learning** algorithm whose aim, like every algorithm implementing this typology of learning, is to **find Hidden Patterns** in unlabeled data. 

\begin{equation}
 \underbrace{\{\mathbf{x}_i\}_{i=1}^N}_{\text{known}} \sim \underbrace{\mathcal{D}}_{\text{unknown}}
\end{equation}

Indeed, we have that, unlike in **Supervised Learning**, any data points is not paired with a label. When we are dealing with **K-Means**, the notion of "Finding Structure" translate into **Deriving a partitioning of the data into Clusters**.

##### FORMALIZING THE TASK:

- **Input**:
    1. $\quad$ A set of points $\{ {x}_1, \ldots, {x_n}\}$ in a d-dim vector space $x \in \R^{d}$, 
    2. $\quad$ $k$ number of clusters to find

- **Output**:
    1. $\quad$ A set of points **"centroids"** $\{ {\mu}_{1}, \ldots, {\mu}_{n}\}$ where ${\mu} \in \R^{d}$, 
    2. $\quad$ A set of **assignment labels** $\{ y_1, \ldots, y_n\}$ that create a **mapping from data points to centroids**, 
    3. The clusters are meant to be disjoint (there is no overlap between clusters).
        - Note that $y$ is a scalar $\in [1,\ldots,k]$;
        - The assignment labels maps a data point to a single centroid;

##### DESCRIPTION OF THE ALGORITHM:

The algorithm can be divided into 3 main steps:
1. **Init of Centroids:** First of all, we need some "reference" centroids to start from. We therefore random sample the $K$ centroids $\{ {\mu}_{1}, \ldots, {\mu}_{n}\}$. A trick that can be used is to random sample from the available points (just choose a few of them).
2. **Assignment Step:** - For every point $i=[1,\ldots,N]$ compute the assignment w.r.t. to the $K$ clusters. 
    - We have to compare $N\times K$ points (indicating: `#Data Points x #Clusters`).
    - Given a point ${x}_{i}$, **we have to find the label of the "closest" centroids in $\{ {\mu}_{1}, \ldots, {\mu}_{n}\}$**
        -  $y_i =  \arg\min_k || {x}_{i} - {\mu}_{k} ||_2^2 $  $\quad\quad$ (Note here the use of ${L}_{2}$)
3. **Update Step:** - Now, it is the **inverse** of before. Now we consider the "centroids".
    - For every centroid $k=[1,\ldots,K]$ **we update its mean based on the previous assignments:**
    - Given a ${\mu}_{k}$, we update it as:  $${\mu}_{k} \leftarrow \frac{\sum_i \delta\{y_{i}=k\} {x}_{i}}{\sum_i \delta\{y_i=k\}}$$




Note that we have: $
\delta_{ij} = \begin{cases}
0 &\text{if } i \neq j,   \\
1 &\text{if } i=j.   \end{cases}
$

It is a formalism to express the concept of:
- DO consider the point in the mean computation if point label corresponds to current centroid.
- DO NOT consider the point in the mean computation if point label DOES NOT corresponds to current centroid.
    



### 1.1.2.1: DISTORTION: ELBOW METHOD:

##### REASONING:
Here, in this section, we are going to apply the **Elbow Method** that is a very popular pre-processing technique built on the idea to run **K-Means Clustering** for a range of clusters k and for each value, we are going to calculate the sum of squared distances from each point to its assigned center, (a.k.a **Distortion**).

We will obtain that the distortions are plotted and the plot looks like an arm then the elbow(the point of inflection on the curve) is the best value of k. Its aim indeed, is to **find the best value for the number of clusters**.

##### RESULT:
From the Plotting of the Elbow Method Result, we have obtained that the "Elbow" we were looking for is found at **K = 2**.

```python
#CREATION OF A NEW DATASET:
df1 = df.copy()                                                     #Here we copy the Processor's Dataset.
df1.drop('data', axis=1, inplace=True)                              #We drop from the copy we have just done, all the rows, except for the 'idle_time%'
df1.drop('interrupt_time%', axis=1, inplace=True)                   #column and the 'processor_time%' one.
df1.drop('priviledged_time%', axis=1, inplace=True)
df1.drop('user_time%', axis=1, inplace=True)
df1.drop('Interrupt/sec', axis=1, inplace=True)
df1.drop('state', axis=1, inplace=True)

#ELBOW METHOD:
distortions = []                                                    #Initializing the list "distortions", that will contain all the Distortion Values in it.
K = range(1,10)                                                     #Setting the range of Ks to try pertaining to a range from 1 to 10.
for k in K:                                                         #Looping over the ks.
    kmeanModel = KMeans(n_clusters=k)                               #Applying the Kmeans with k number of clusters.
    kmeanModel.fit(df1)                                             #Fit the data in the K-Means with K centroids.
    distortions.append(kmeanModel.inertia_)                         #Compute and append the Distortion value in the distortions list.
    
#PLOTTING THE ELBOW METHOD RESULT:
plt.plot(K, distortions, 'bx-')                                     #Plotting the Elbow Methos, with K Clusters and the Distortions.
plt.xlabel('k')                                                     #Inserting the x-axis label.
plt.ylabel('Distortion')                                            #Inserting the y-axis label.
plt.title('Distortion: The Elbow Method')                           #Inserting the title in the plot.
plt.show()                                                          #Showing the result.
```

### 1.1.2.3: K-MEANS CLUSTERING: RANDOM INITIALIZATION (BY-HAND):

In this subsection, we have developed the **K-Means Clustering with Init Initialization by Hand**, namely following the explanation given at the starting of point 1.2. (With step 0, 1, 2)

```python
#SETTING UP THE SEED:
np.random.seed(42)                                                      #We set up the seed in order to guarantee reproducibility.   

#CREATION OF A NUMPY ARRAY FROM OUR DATAFRAME:                          #We need to do this due to some inconsitency in the type of the data.
df_final_x = pd.DataFrame(df['idle_time%'])                             #Creation of a Dataframe with the 'idle_time%'.                           
df_final_y = pd.DataFrame(df['processor_time%'])                        #Creation of a Dataframe with the 'processor_time%'. 
df_final_x = pd.Series.tolist(df_final_x)                               #Converting to NumPy array our 'idle_time%' Dataframe. 
df_final_y = pd.Series.tolist(df_final_y)                               #Converting to NumPy array our 'processor_time%' Dataframe. 
df_final = np.column_stack((df_final_x,df_final_y))                     #Creation of a Dataframe with the 'idle_time%' and 'processor_time%' columns. 

#STEP 0:
def init_kmeans(X, k):                                                  #Declaring a funciton for our 0 Step.
    N, dim = X.shape                                                    #Recover the Dimensions of our Dataframe.
    idx = np.random.choice(N, k)                                        #We select k samples from the points we have randomly. Note that actually the indexes are random sampled. 
    return X[idx]                                                       #Since we random sample the indexes, we return the k-random sampled points.

#STEP 1:
def get_assignments_kmeans(X, centers):                                 #Declaring a funciton for our 1 Step.
    P = X[:, np.newaxis, :]                                             #We have to find the label of the "closest" centroids (Nx1xD).
    diff_sq = (P - centers)**2                                          #Computing the L2 Loss. (Nx1xD - KxD -> NxKxD).
    y = diff_sq.sum(axis=2).argmin(axis=1)                              #L2 Loss applied.
    return y                                                            #Return all the assignments. Y will have an Nx1 Dimensionality.

#STEP 2:
def update_centers_kmeans(X, y, K):                                     #Declaring a funciton for our 2 Step.
    # Select the points to the k center and remove N dimension with mean over axis 0.
    return np.array([X[y == k].mean(axis=0) for k in range(K)])         #Note that the "==" corresponds to the Delta(i,j) in the formal formula. It corresponds, to the Indicator Function.

#DOUBLE CHECK TO CONTROL WHETHER THE LABELS ARE EQUAL:
centers = init_kmeans(df_final, 2)                                      #Initialize the Centroids with two random numbers.  
y_assignments = get_assignments_kmeans(df_final, centers)               #Compute the assignments using the function built above.
labels = pairwise_distances_argmin(df_final, centers)                   #Compute the assignments using the built-in function.
assert np.array_equal(labels, y_assignments), 'Assignment not correct!' #Check if the result obtained is correct or not.
```

```python
#WRAPPING UP ALL THE ALGORITHM:
def kmeans_byhand(X, n_clusters):                                       #Declaring the KMeans function taking in input a Dataset and a number of clusters.                                 
    np.random.seed(42)                                                  #We set up the seed in order to guarantee reproducibility. 
    iter = 1                                                            #Setting up a counter for counting the number of iterations the algorithm will do.
    
    #STEP 1: RANDOM INIT OF THE CENTROIDS:
    centers = init_kmeans(X, n_clusters)                                #Calling the previously contructed function for the Step 1.
    
    while True:                                                         #Entering the Loop, and therefore entering step 2 and step 3 of the KMeans algorithm.
        
        #STEP 2: ASSIGNMENT STEP:
        y_assignments = get_assignments_kmeans(X, centers)              #Calling the previously contructed function for the Step 2.
        
        #STEP 3: UPDATE STEP:
        new_centers = update_centers_kmeans(X, y_assignments, n_clusters)#Calling the previously contructed function for the Step 3.
        
        #CHECK FOR CONVERGENCE:
        if np.allclose(centers, new_centers):                           #This is a basic implementation of this step. A more advanced one would be to check for if the loss is plateauing.
            #print(f'KMeans terminates after the Iteration number: {iter}')#Print out the number of iteration at which the algorithm stops.
            break                                                       #The "np.allclose" function returns True if two arrays are element-wise equal within a tolerance. Break the while if it is so.
        
        #UPDATE CENTROIDS FOR THE NEXT ITERATION:
        centers = new_centers                                           #Update the Centroids' new value. 
        iter += 1                                                       #Increase the Iteration number.
    
    return centers, y_assignments                                       #Return from KMeans both the centroids and the labels.

#CALLING THE KMEANS BY-HAND IMPLEMENTATION:
centers_kmeans, labs_kmeans_byhand = kmeans_byhand(df_final, 2)                #Obtainig the centroids and labels of the KMeans.

#PLOTTING THE KMEANS BY-HAND IMPLEMENTATION RESULT:                
plt.scatter(*df_final.T, c=labs_kmeans_byhand, s=40, cmap='viridis')    #Plotting the Dataset's Samples, colored thanks to Labeling.
plt.scatter(*centers_kmeans.T, s=250, marker='*', c='red', label='centroids')  #Plotting also the Centroids for every K-Means centroid we have. 
plt.xlabel('processor_time%')                                           #Inserting the x-axis label.
plt.ylabel('idle_time%')                                                #Inserting the y-axis label.
plt.title('KMeans By Hand Resolution')                                  #Inserting the title in the plot.
plt.legend()                                                            #Inserting the Legend.
plt.show()                                                              #Showing the result obtained.
```

### 1.1.2.4: K-MEANS CLUSTERING: RANDOM INITIALIZATION AND ELKAN ALGORITHM:

##### LLOYD ALGORITHM DRAWBACKS:
There are several drawbacks in using k-means clustering, such as the **Sensitivity to Outliers**, the problem of **how to choose k** and the problem on **how to initialize the k centers**. 

For what concerns **How to Choose K and how to Initialize the Centroids**: Those two are solved by using the **Elbow Method** and **K-Means++** respectively. 

In practice, we have that k-means is often run multiple times with different initial centers to get good results. For such approaches, the runtime of k-means, especially in high dimensions with a large set of samples ,is crucial. We have, indeed, that the **Running Time** for the **Lloyd's Algorithm** is equal to: **$\mathcal{O}(wnkd)$** for w iterations, k centers and n points in d dimensions.

What the **Elkan Algorithm** proposes is to use the **Triangle Inequality** in order to reduce the number of distance calculations and, thus, reduce runtime. Lloyd's Algorithm spends a lot of processing time computing the distances between each of the k cluster centroids and the n data points. Since points usually stay in the same clusters after a few iterations, much of this work is unnecessary, making the Naive implementation very inefficient.

What the **Elkan Inequality's Traingle Inequality** claims is the following:
##### CLAIM:
If center ${c'}$ is close to point ${x}$ and some other center ${c}$ is far away from another center ${c'}$, then ${c'}$ must be closer than ${c}$ to ${x}$.

## 1.1.3: K-MEANS++ CLUSTERING:

##### K-MEANS LIMIT:

One main problem that really slow down the **K-Means Algorithm** is in the **choice on the initial centroids value**. The basic and Naive idea used by the K-Means, is to make use of the **Random Initializatio**, choosing at random points in the list of points we have. 

##### K-MEANS++ IDEA: FURTHEST-FIRST HEURISTIC:

A very great idea that arised and that is used by the **K-Means++ Algorithm** is in the **Semi-Random Initialization** by picking initial means as far from each other as possible. This choice takes the name as **Furthest-First Heuristic**. What makes this Heuristic **Semi-Random** is in the fact that the **First Choice** for choosing the First Centroid is still **Random**. Then, all the others are chosen deterministically. Indeed, the **Furthest-First Heuristic** works as:
1. Choose  ${\mu}_{1}$ arbitrarily between the data points available.
2. For  $k=[2,\ldots,N]$ we do:
    - Find the data point ${x}_{m}$ that is as far as possible from all previously selected means $[{\mu}_{1},\ldots,{\mu}_{k-1}]$
    - Assign  ${\mu}_{k} \leftarrow {x}_{m}$,  where we have that $$m ={\arg\max_m {\big( \min_{k<k^{\prime}} \left| \left| {x}_m - {\mu}_k^{\prime}\right|\right|_2^2 \big)}}$$


##### K-MEANS++ ALGORITHM:

We can therefore sum up the **K-Means ALgorithm** as:
1. Choose ${\mu}_1$ arbitrarily between data points
2. For $k=[2,\ldots,K]$:
    - **Inverse Transform Sampling** wrt to distances between centroids, namely Transform the closed set $D$ of distances into a distribution (probability mass function - pdf) as $$p(d^{\prime})= \frac{d^{\prime}}{\sum{d \in D} d}$$
    - $Pr[{\mu}_k={x}_m] \propto \min_{k<k^{\prime}} \left| \left| {x}_m - {\mu}_k^{\prime}\right|\right|_2^2 \qquad$

3. Repeat until convergence Lloyds method:
    - **Assignment step:** $$\forall i\in[1,N] \quad y_i =  \arg\min_k || {x}_i -{\mu}_k ||_2^2 $$
    - **Update step:** $$\forall k\in[1,K] \quad  {\mu}_k \leftarrow \frac{\sum_i \delta\{y_i=k\}{x}_i}{\sum_i \delta\{y_i=k\}}$$


##### K-MEANS++: A FAMILIY OF ALGORITHMS:
K-means++ opens the possibilty of interpret K-means as a **Familiy of Algorithms**. Indeed, K-means++ is also called Kmeans with $D^2$ sampling because we sample proportionally to Euclidean norm $\ell_2$. 
You can think of sampling wrt to:
 - $\ell_0$ "norm" (total number of non-zero element in a vector) $\rightarrow$ **Random Sampling**. 
 - $\ell_2$ norm $\rightarrow$ when we have **K-means++**.
 - $\ell_{\infty}$ norm $\rightarrow$ then we have **K-Means with Furthest-First Heuristic** (we take max of distances).

### 1.1.3.2: K-MEANS CLUSTERING: K-MEANS++ (BY-HAND):

In this subsection, we will try to develop the **K-Means++ Algorithm** all **By-Hand**, following the description we've done above.

```python
#SETTING UP THE SEED:
np.random.seed(42)                                                      #We set up the seed in order to guarantee reproducibility.   

#SETTING UP THE NUMBER OF CENTROIDS:
N_centers = 2

#STEP 0:
def init(X, K, plusplus=True):                                          #Definition of the "Init" function that will be the one that handles the Initialization, both of the first centroid and both after.
    N, dim = X.shape                                                    #Taking the number of samples and its dimension from the Dataset.
    
    #FIRST CENTROID'S INITIALIZATION:
    if not plusplus:                                                    #When we are in the first initialization of the First Centroid.                                   
        idx = np.random.choice(N, K)                                    #We select k samples randomly from the points we have. 
        return X[idx]                                                   #Return the index of the sample we have randomly choosen as first Centroid.
    
    #OTHERS CENTROID'S INITIALIZATION:
    else:
        idx_first = np.random.choice(N, 1)                              #We select a point.
        centers = X[idx_first]                                          #Initialize the centers. 
        dists = distance(X, centers)                                    #Initialize the distances.
        for k in range(1, K):                                           #Entering the loop from 0 to k-1.
            min_dist = dists.min(axis=1)                                #Compute min across all k-centers, for all points in our Dataset. We have that 'dists' is NxK so min_dist will be Nx1.
            idx = inverse_sampling(min_dist/min_dist.sum(), 1)          #Select one of the N proportional to higher distance.
            new_center = X[idx]                                         #Take its value using the index we are gien above.
            dist_new = distance(X, new_center)                          #Compute the distance between all points to the new center.
            centers = np.vstack([centers, new_center])                  #Update pool of centers.
            dists = np.hstack([dists, dist_new])                        #Update pool of distances.
        return centers                                                  #Return the centers that are computed after the loop.
    

#STEP 1.1:
def inverse_sampling(pmf, n_samples):                                   #Definition of the "Inverse_Sampling" function that will be the one that transforms the closed set D of distances into a distribution
                                                                        #(probability mass function - pdf).
    U = np.random.rand(n_samples)                                       #Take at random n samples.
    sampled_idx = np.argmin((U[:, None] > pmf.cumsum()), axis=1)        #Apply the argmin function in such a way to get the index of the minimum point in terms of distances.
    return sampled_idx                                                  #Return its index.

#STEP 1.2:
def distance(X, centers):                                               #Definition of the "Distance" function that will be the one that that computes the distance between a point and the centroids.
    P = X[:, np.newaxis, :]                                             #Take the Nx1xD data.
    diff_sq = (P - centers)**2                                          #Compute the L2 norm. (Nx1xD - KxD -> NxKxD)
    diff_sq_sum = diff_sq.sum(axis=2)                                   #Compute the square of the L2 Norm. 
    return diff_sq_sum                                                  #Return the sum of the L2 Norm.

#STEP 2.1:
def get_assignments(X, centers):                                        #Definition of the "Get_assignment" function that will be the one that gets the new assignments(labels).
    dist_sq = distance(X, centers)                                      #Get the distance from the points to centroids. 
    y = dist_sq.argmin(axis=1)                                          #Select argmin distance.
    return y, dist_sq.min(axis=1)                                       #Return all the assignments. Y will have an Nx1 Dimensionality.

#STEP 2.2:
def update_centers(X, y, K):                                             #Definition of the "Update_Centers" function that will be the one that handles the Update of centers during our Iterations.
    return np.array([X[y == k].mean(axis=0) for k in range(K)])          #Note that the "==" corresponds to the Delta(i,j) in the formal formula. It corresponds, to the Indicator Function.

#DOUBLE CHECK TO CONTROL WHETHER THE LABELS ARE EQUAL:
centers = init(df_final, N_centers)                                      #Initialize the Centroids with two random numbers.  
y_assignments, _ = get_assignments(df_final, centers)                    #Compute the assignments using the function built above.
labels = pairwise_distances_argmin(df_final, centers)                    #Compute the assignments using the built-in function.
assert np.array_equal(labels, y_assignments), 'Assignment not correct!'  #Check if the result obtained is correct or not.
```

```python
#WRAPPING UP ALL THE ALGORITHM:
def kmeans_plus_byhand(X, n_clusters):                                  #Function that will merge all the algorithm in one big function.
    np.random.seed(42)                                                  #We set up the seed in order to guarantee reproducibility. 
    iter = 1                                                            #Setting up a counter for counting the number of iterations the algorithm will do.
    
    #STEP 1: RANDOM INIT OF THE CENTROIDS:
    centers = init(X, n_clusters)                                       #Calling the previously contructed function for the Step 1.
    loss = []                                                           #Declare a Loss List for keeping track of the loss we have.
    
    while True:                                                         #Entering the Loop, and therefore entering step 2 and step 3 of the KMeans algorithm.
        
        #STEP 2: ASSIGNMENT STEP:
        y_assignments, dist_sq = get_assignments(X, centers)            #Calling the previously contructed function for the Step 2.
        
        #COMPUTE LOSS:  
        loss.append(dist_sq.mean())                                     #Compute the distance between points and center, average over points then.
        
        #STEP 3: UPDATE STEP:
        new_centers = update_centers(X, y_assignments, n_clusters)      #Calling the previously contructed function for the Step 3.

        #CHECK FOR CONVERGENCE:
        if len(loss) > 2 and np.allclose(loss[-2], loss[-1], atol=1e-5):# If two last losses are close enough we stop.
            break
        
        #UPDATE CENTROIDS FOR THE NEXT ITERATION:
        centers = new_centers                                           #Update the Centroids' new value. 
        iter += 1                                                       #Increase the Iteration number.
        #print(f'> Iteration {iter}: loss {loss[-1]}')                  #Print ou the Loss we have at the i-th iteration.

    return centers, y_assignments                                       #Return from KMeans both the centroids and the labels.


#CALLING THE KMEANS++ BY-HAND IMPLEMENTATION:
centers, labs_kmeans_plus_byhand = kmeans_plus_byhand(df_final, N_centers)#Obtainig the centroids and labels of the KMeans.



#PLOTTING THE KMEANS BY-HAND IMPLEMENTATION RESULT:                
plt.scatter(*df_final.T, c=labs_kmeans_plus_byhand, s=40, cmap='viridis')#Plotting the Dataset's Samples, colored thanks to Labeling.
plt.scatter(*centers.T, s=250, marker='*', c='red', label='centroids')  #Plotting also the Centroids for every K-Means centroid we have. 
plt.xlabel('processor_time%')                                           #Inserting the x-axis label.
plt.ylabel('idle_time%')                                                #Inserting the y-axis label.
plt.title('K-Means++ By Hand Resolution')                               #Inserting the title in the plot.
plt.legend()                                                            #Inserting the Legend.
plt.show()                                                              #Showing the result obtained.
```

### 1.1.4.1: AIC AND BIC EVALUATIONS ON GMM:

In this subsection, we find a way to evaluate the work done by MoG on our Dataset. We therefore choose to use **Information criterion**. The **Akaike Information Criterion (AIC)** and the **Bayesian Information Criterion (BIC)** provide **measures of model performance** that account for model complexity. AIC and BIC combine a term reflecting how well the model fits the data with a term that penalizes the model in proportion to its number of parameters. We have that:
- **AIC**: It relies on an asymptotic approximation that may not hold for a given finite data set.
$$ \text{(AIC)}  \qquad \arg\min_K \mathcal{L}_K  + \underbrace{K\cdot 2D}_{\text{regularization}}$$

- **BIC**: It relies on the assumption that the model errors are independent and normally distributed. 
$$ \text{(BIC)} \qquad \arg\min_K \mathcal{L}_K  + \underbrace{K\cdot\log(D)}_{\text{regularization}}$$


Both AIC and BIC are functions of the parameter count and the maximized likelihood, that is, the probability of the data given the maximum-likelihood fit of the model. Counting parameters is not in general a good method of estimating model complexity. Nevertheless, where applicable, AIC and BIC provide a quick and easy way to compare models.

## 1.1.5: COMPARISON REALITY, GMM, K-MEANS, KMEANS ++:

In this section we are going to have a Discussion about the comparison of the two Methods: **K-Means VS GMM**. 

##### ADVANTAGES AND DISADVANTAGES:
First and foremost though, let's do a distinction between **Advantages and Disadvantages** of the two main Methodology.
1. **K-Means:**
    - **Advantages:**
        - Simple to understand.
        - Very quick (all that is being computed is the distance between each point and cluster center).
        - Easy to implement

    - **Disadvantages:**
        - Have to choose K manually.
        - Depends on initial centroid locations.
        - Potential to misrepresent centroid positions due to outliers.

2. **GMM:** 
    - **Advantages:**
        - Can analyze more complex and mixed data w.r.t. K-Means.
        - Can handle outliers more easily.

    - **Disadvantages:**
        - More difficult to directly interpret results.
        - Does not directly assign data points to clusters.

##### SIMILARITIES AND DIFFERENCES:
- **SIMILARITIES:**
    - **Starting Points of Centroids Selection:** This is very important as it can have a big effect on the outcome. Various methods have been proposed to initialize the cluster centers. Common methods are:
        - **Forgy:** (chooses K data points randomly and use those as the initial centroid locations)
        - **Random Partition:** (randomly assigns a cluster to each data point and then calculates a centroid based on this clustering- this generally creates initial centroids close to the middle of the data set)
        - **K-Means++:** (chooses initial centroids that are as far apart from each other as possible, effectively around the edges of the data set).


    - **Minima: Tendency towards Local Minima rather than Global Minimum:** Because the initialization of the centroids is essentially a guess, they can start far away from the true cluster centers in the data. The two methods always converge (meaning they will always find a solution) but not necessarily the optimal one. On the way, they may fall into a local minimum get stuck there.

    ![title](img/globalvslocal.png)

    
- **DIFFEREENCES:**
    -**Approach: Deterministic vs Probabilistic (HARD-CLUSTERING VS SOFT-CLUSTERING):** K-Means and GMM have a completely different approach w.r.t. their approach to problems:
        - **K-Means:** It uses a **Deterministic Approach** and assigns each data point to one unique cluster. This is referred to as a **Hard Clustering Method**.
        - **GMM:** It uses a **Probabilistic Approach** and gives the probability of each data point belonging to any of the clusters. This is referred to as a **Soft Clustering Method**.

    - **Updating Centroids:**
        - **K-Means:** It uses only the data points assigned to each specific cluster to update the centroid mean. 
        - **GMM:** It uses all data points within the data set to update the centroid weights, means and covariances. 
    
    - **Feature Standardization:**
        - **K-Means:** The circular distance measure around a centroid can make feature standardization necessary if one or more of the dimensions will dominate the calculation.
        - **GMM:** It automatically takes the issue into account by its calculation and use of the covariance matrix.


## 1.1.6: SILHOUETTE COEFFICIENT:

##### INTRODUCTION:
The **Silhouette Score** is used to evaluate the **Quality of Clusters** created using clustering algorithms such as K-Means in terms of how well samples are clustered with other samples that are similar to each other. It is used to measure **how dense and well-separated the clusters are**.

The Silhouette score is calculated for each sample of different clusters. In order to calculate the Silhouette score for each observation / data point, the following distances need to be found out for each observations belonging to all the clusters:

 Mean distance between the observation and all other data points in the same cluster. This distance can also be called as **Mean Intra-Cluster Distance**. The mean distance is denoted by ***a***.

 Mean distance between the observation and all other data points of the next nearest cluster. This distance can also be called as **Mean Nearest-Cluster Distance**. The mean distance is denoted by ***b***.

Silhouette score, S, for each sample is calculated using the following formula:

$$S = \frac{(b-a)}{max(a,b)} = \frac{Mean Nearest-Cluster Distance - Mean Intra-Cluster Distance}{max(Mean Intra-Cluster Distance, Mean Nearest-Cluster Distance)}$$

The value of Silhouette score varies from -1 to 1. We have that:
- If the score is 1, the cluster is dense and well-separated than other clusters. 
- A value near 0 represents overlapping clusters with samples very close to the decision boundary of the neighbouring clusters.
- A negative score [-1, 0] indicate that the samples might have got assigned to the wrong clusters.

### 1.1.6.2: SILHOUETTE PLOT:  

The **Silhouette Plot** can be used to select the most optimal value of the K (no. of cluster) in K-means clustering. It can be used in alternative to the more commonly known **Elbow Method**.
The aspects to look out for in the Silhouette Plot are the following: 
- Cluster scores below the average Silhouette Score.
- Wide fluctuations in the size of the clusters.
- Thickness of the Silhouette Plot.

From our example, it turns out that: 
- The value of ***n_clusters*** as ***3***, ***4*** and ***5*** looks to be suboptimal for the given data due to the following reasons:
    - Presence of Clusters with below Average Silhouette Scores. 
    - Wide fluctuations in the size of the Silhouette Plots.

The value of ***2*** for n_clusters looks to be **Optimal One**. Here indeed, the Silhouette Score for each cluster is above the Average Silhouette Scores, even if the green one is a little bit below. Also, the fluctuation in size is similar, even if the Blue Cluster is a little bit big w.r.t. the green one. The thickness of the silhouette plot representing each cluster also is a deciding point, in which it seems to be more uniform than the other plots. Thus, one can select the optimal number of clusters as 2.

After doing this, we can proceed by scaling the Accuracy at the same level of the the **Time Performances**, namely, we standardize it, by divinding it by 1000. Then, we are going to take this last result and add it to the Time Performances. The best algorithm to be choosen would be the one to have the less **Time+Accuracy Value**.

#### RESULT:

## 2.1.3 : PARAMETRIC BOOTSTRAP PAGE FAULTS' MEAN 

Using Parametric Bootstrap, we compute the average number of Page Faults occured during the execution of a Python Program (see above).
For the Bootstrapped version, we opt for the skewnormal distribution that seems to fit the data quite well. 

```python
#PLOTTING THE DISTRIBUTION:
plt.hist(df_memory['Page_Faults/sec'], density = 1, facecolor='purple',                     #We plot wind as an Histogram Distribution.
         edgecolor = 'white', alpha=0.5, bins = 20)  
plt.title("DISTRIBUTION OF Page_Faults/sec")                                                #Insertion of the title in the Plot.                                                               .
plt.xlabel('Page_Faults/sec')                                                               #We add up the x-axis label.
plt.ylabel('Density')                                                                       #We add up the y-axis label.
plt.show()                                                                                  #We show the plot.

#MLE DEFAULT METHOD:
hat_a, hat_loc, hat_scale = st.skewnorm.fit(df_memory['Page_Faults/sec'])                    
print('hat_a',hat_a,'hat_loc',hat_loc,'hat_scale',hat_scale)                                #We obtain the  actual parameters of the skewnorm distribution using the fit.
print(hat_a, hat_loc, hat_scale)                                                            #We print out the three parameters of our new log distribution under a skewnorm model.

#PDF OF THE SKEWNORM DISTRIBUTION:
rv_fit  = st.skewnorm(a = hat_a, loc = hat_loc, scale = hat_scale)                          #We create the distribution with previously found parameters.
v_grid  = np.linspace(df_memory['Page_Faults/sec'].min(),                                   #Found parameters.   
                      df_memory['Page_Faults/sec'].max(), num = 501)                                       
pdf_val = rv_fit.pdf(v_grid)                                                                #We call the pdf function over it and assign that to the variable 'pdf_val'.
#fitted model 

#1-st PLOT: PAGE-FAULT DISTRIBUTION:
                           
plt.hist(df_memory['Page_Faults/sec'], density = 1, facecolor='purple',                     #Plotting our Histogram Distribution.
         edgecolor = 'white', alpha=0.5, label = "Data") 
plt.plot(v_grid, pdf_val, '#fdc0cc', linewidth = 5,label = "Estimated PDF")                 #Plotting our our Estimated PDF.
plt.title("DISTRIBUTION OF Page_Faults/sec")                                                #Insertion of the title in the Plot.   
plt.xlabel('x')                                                                             #We add up the x-axis label.
plt.ylabel('PDF')                                                                           #We add up the y-axis label.
plt.legend(loc='upper left')                                                                #We add up the Legend of the plot.
plt.show()                                                                                  #We show up the plot.

# 2nd PLOT: THE SKEWNORMAL PDF OVER OUR DISTRIBUTION FIT: QUANTILE-QUANTILE-Plot(QQ-Plot):
stmod.qqplot(data = df_memory['Page_Faults/sec'], dist = st.skewnorm(hat_a, hat_loc, hat_scale), line='45', fmt='b')#Let's buiild up a QQ-Plot to see how the SkewNormal Distribution 
                                                                                            #fits ours.
plt.show()                                                                                  #Show the qqplt.
```

Assume (again) that the parameter of interest $\tau$ is indeed the population mean $\mathbb{E}(X)$. <br>
It is a function of the "technical" parameter $\boldsymbol{\theta} = (\mathrm{loc},\mathrm{scale},\mathrm{shape})$:
$$\tau = \mathbb{E}(X) = h(\boldsymbol{\theta}) = \mathrm{loc} + \mathrm{scale}\times\delta\times\sqrt{\frac{\pi}{2}}$$  
where $$\delta = \frac{\mathrm{shape}}{\sqrt{1+{\mathrm{shape}^2}}}$$
Hence, by the **equivariance** of MLE, we know that
$$\widehat{\tau}_{\tt ML} = h\big( \widehat{\boldsymbol{\theta}}_{\tt ML}\big) = \widehat{\mathrm{loc}}_{\tt ML}+\widehat{\mathrm{scale}_{\tt ML}}\times\widehat{\delta}_{\tt ML}\times\sqrt{\frac{\pi}{2}} $$
In fact,

```python
#We are writing hat_a,hat_loc,hat_scale in terms of delta,epsilon,
#omega according to the paramters of the skew normal distribution (just for simplicity) 
epsilon = hat_loc                                   #It represents the shifting parameter of the skew-normal distribution
omega = hat_scale                                   #It represents how much the distribution is spread (The larger the scale parameter, the more spread out the distribution.) 
delta = hat_a / np.sqrt(1+(hat_a)**2)               #It must affect the shape of a distribution rather than simply shifting it (as a location parameter does) or stretching/shrinking it (as a scale parameter does).
hat_tau = epsilon + omega*delta*(np.sqrt(2/np.pi))  #We compute the estimated value of the population mean,by using the equivariance of the MLE.
print(hat_tau)                                      #We print it out
```

```python
#PARAMETRIC BOOTSTRAPPED SOLUTION 
np.random.seed(0)                                                   #We fix the seed to allow reproducibility.
B = 500                                                             #Size of the bootstrap.
boot_stat = np.zeros(B)  
for b in range(B):                                                  #We initialize our boot_stat as an array filled with zeros.
    x_bt = rv_fit.rvs(len(df_memory['Page_Faults/sec']))            #Sample for the fitted population.
    a_bt, loc_bt, scale_bt = st.skewnorm.fit(x_bt)                  #Obtain the bootstrap replicates of the MLEs.
    eps_bt = loc_bt                                                 #We write a_bt,loc_bt,scale_bt in terms of delta_bt,eps_bt,omega_bt just for simplicity (same as before).
    omega_bt = scale_bt
    delta_bt = a_bt / np.sqrt(1+(a_bt)**2) 
    boot_stat[b]  = eps_bt + omega_bt*delta_bt*(np.sqrt(2/np.pi))  #Use equivariance to get the bootstrap replicates for the MLEs of the parameter of interest.

#BOOTSTRAPPING ESTIMATION OF THE MEAN: 
parametric_b_mean = boot_stat.mean()                               #Given our boot_stat distribution we extract the mean.
print(parametric_b_mean)
```

#### INTRODUCTION:

#### OBJECTIVE :

- **SUPPORTING TOOLS**: As support Tools, we performed an **ANOVA** global test, assuming the average number of Cache Faults is equal for all our PCs. Then, A **t-test** was performed pairwise,checking whether the means of the two groups are equal.

- **2.2.2: PAIRWISE T-TEST**

It uses the following null and alternative hypotheses:

- H0 (NULL HYPOTHESIS): 1 = 2 = 3 =  = k (all the population means are equal)

### 2.2.2 : PAIRWISE T-TEST

After having performed a global test, we are interested in the average number of Cache Faults for each pair of PCs. 
A two-sample T-test allows us to conclude that only a pair of PC "has the same behaviour". In the other cases, the null hypothesis is always rejected (as the Anova result)

```python
#Computing the variance for each pair of Cache Faults' rates 
print(np.var(df_memory_a['Cache_Faults/sec']), np.var(df_memory_m['Cache_Faults/sec']))     
print(np.var(df_memory_m['Cache_Faults/sec']), np.var(df_memory_g['Cache_Faults/sec']))
print(np.var(df_memory_a['Cache_Faults/sec']), np.var(df_memory_g['Cache_Faults/sec']))

#Since the value of the variance is greater than 4, we can conclude that they have equal variance.

#Performing a pairwise t-test 
t_test_am = stats.ttest_ind(a=df_memory_a['Cache_Faults/sec'], b=df_memory_m['Cache_Faults/sec'], equal_var=True)   
t_test_mg = stats.ttest_ind(a=df_memory_m['Cache_Faults/sec'], b=df_memory_g['Cache_Faults/sec'], equal_var=True)
t_test_ag = stats.ttest_ind(a=df_memory_a['Cache_Faults/sec'], b=df_memory_g['Cache_Faults/sec'], equal_var=True)
print(t_test_am, t_test_mg, t_test_ag )
```

### 4.1.3 : COMPUTING LEVERAGE AND COOK DISTANCES 

We want to look at the most influential points,i.e points that have greater potential influence on the least squares estimates. 
For an observation to actually be influential, it must have both a relatively large leverage and a relatively large residual.
An observation that falls far from the mean on the explanatory variables has a relatively large leverage.

Let's start by computing the **Leverage**.
$$Leverage_i = hi = \frac{1}{n} + \frac{(x_i - \bar{x})^2}{\sum_{k=1}^{n} (x_k - \bar{x})^2}$$ 

```python
#CALCULATING X's AVERAGE: 
x_avg=np.mean(df_cache['Copy_Reads/sec'])                    #Computing the average of the X column.

#COMPUTING THE LEVERAGE:
lev_list=[]                                                  #Declaring a list of Leverages.
n=len(df_cache['Copy_Reads/sec'])                            #Computing the length of the Dataset.
denominator=0                                                #Setting the denominator variable to zero.
for el in df_cache['Copy_Reads/sec']:                        #For each element in the Copy_Reads/sec column.
    denominator+= (el+x_avg)**2                              #Computing the sum of the (element + avg)**2.

for el in df_cache['Copy_Reads/sec']:                        #For each element in the Copy_Reads/sec column.
    lev=(1/n)+((el-x_avg)**2/denominator)                    #Compute the entire leverage for each element.
    lev_list.append(lev)                                     #Append the result obtained.

print(sorted(lev_list, reverse = True))                      #Printing out the Leverage List obtained sorted.
```

Cooks distance is a diagnostic that uses the leverage and the residual, based on the change in ${\hat{\beta}}_{j}$ when the observation is removed from the data set. For the residual $e_i$ and leverage $h_i$ for observation i, Cooks distance is: $${D}_{i} = \frac{{e^2}_{i}{{h}_{i}}}{(p+1)s^2(1-{h}_{i})^2}$$ 
where ${s}^2$ is an estimate of the conditional variance ${\sigma}^2$ and $p$ : is the number of Explanatory Predictors, (in our case $p = 1$).

```python
#COMPUTATION OF BETA_0 AND BETA_1:
y_avg=np.mean(df_cache['Data_Maps/sec'])                                                    #Computing yhat from the response variable Y.
num = np.sum((df_cache['Copy_Reads/sec'] - x_avg) * (df_cache['Data_Maps/sec'] - y_avg))    #Computing the numerator of Beta_1.
den = np.sum((df_cache['Copy_Reads/sec'] - x_avg)**2)                                       #Computing the numerator of Beta_1.
beta_1 = num/den                                                                            #Computing Beta_1.
beta_0 = y_avg - (beta_1*x_avg)                                                             #Computing Beta_0.

#COMPUTING THE MU HAT:
mu_list=[]                                                                                  #Declaring a mu hat list. 
for i in range(n):                                                                          #For every element in the length.
    mu_list.append(beta_0+beta_1*df_cache['Copy_Reads/sec'][i])                             #Append the muhat result to the list.

#COMPUTING THE RESIDUAL LIST e_i:
e_list=[]                                                                                   #Declaring a residual list. 
for i in range(n):                                                                          #For every element in the length.
    e_list.append((df_cache['Data_Maps/sec'][i]-mu_list[i]))                                #Append to the residual list the value we have obtained.

#COMPUTING S_SQUARED:
s=0                                                                                         #Declaring an s variable.
for i in range(n):                                                                          #For every index in the range of the length.
    s+=e_list[i]**2                                                                         #Sum up to the s variable the residual_i squared.
s_square=s/(n-2)                                                                            #Compute the whole S^2, by dividing by n - (p + 1), where p, in our case, is 1.

#CALCULATING COOK'S DISTANCE
cook_dist_list=[]                                                                           #Declaring a Cook_Distance List.
for i in range(n):                                                                          #For every index in the range of the length.
    cook_dist_list.append(((e_list[i]**2)*(lev_list[i]))/                                   #Computing the Cook's Distance element and append it to the list.
                          (2+s_square)*(1-lev_list[i])**2)

print(sorted(cook_dist_list, reverse = True))                                               #Printing out the list in descending order.
```

## 5.1.2 : CORRELATION MATRIX 

We compute the correlation matrix for each possible pair of features in the dataset.
It will tell what are the most and lowest correlated features. 

```python
#SETTING THE SEED:
np.random.seed(42)                                                              #Setting the Seed in order to guarantee the Reproducibility of our Code.

#OBTAINING DATASETS'COLUMNS HEAD VALUES:
h_list=df_conf.columns.values                                                   #Saving the Dataframe Columns values in a list.

#DEFINING D, THE LENGTH OF THE H_LIST:
D = len(h_list)                                                                 #Setting the D value, the length of the list.

#DEFINING OUR FUNCTION THAT WILL COMPUTE THE CIs:
def compute_corr(h_list,dataset):                                               #Defining a function in order to use it for both the Datasets.
    corr_matrix=np.zeros(shape=(D,D))                                           #Define a Correlation Matrix where j,k entry is the correlation of j-th row and kth column.
    for j in range(D):                                                          #Looping over the cells.
        for k in range(j,D):                                                    #Looping from i to j in such a way to loop only throughout half of the Matrix(since it is mirrored).
            if j!=k:                                                            #Checking whether we are not in the diagonal of our Matrix.
                col_1=h_list[j]                                                 #Taking only the column j name.
                col_2=h_list[k]                                                 #Taking only the column k name.
                new_df=dataset[[col_1,col_2]]                                   #Creating a new Dataseset only with this two columns.

                #NON-PARAMETRIC BOOSTRAPPING:
                B = 1000                                                        #Setting up the number "B" that we will use in Boostrapping.
                boot_stat = np.zeros(B)                                         #Declaring a vector of zeros.
                for b in range(B):                                              #Looping over the B amount.
                    x_boot = new_df.sample(n = len(new_df), replace = True)     #A vector with samples from our interest values (with replacement).
                    boot_stat[b] = x_boot.corr().iloc[0,1]                      #Compute the correlation Matrix of that vector, that is inserted in boot_stat, and take the Correlation 
                                                                                #value.
                #CONFIDENCE INTERVAL COMPUTATION:
                hat_rho = np.abs(new_df.corr().iloc[0,1])                       #We compute the absolute value of the estimator of the Pearson Correlation Coefficient.                     
                corr_matrix[j,k]=hat_rho                                        #Insert the value of the the Correlation for a couple of j,k columns in the Correlation Matrix.
                corr_matrix[k,j]=hat_rho                                        #Insert the value of the the Correlation for a couple of j,k columns in the Correlation Matrix, in indexes
                                                                                #that are Mirrored.
            else:
                corr_matrix[j,k]=1                                              #If we are in the Diagonal Matrix, we know that the Correlation is of course 1.
    return corr_matrix                                                          #Returning the Correlation Matrix. 

#OBTAINING THE CONFIDENCE INTERVAL MATRICES:
corr_matrix=compute_corr(h_list,df_conf)                                        #We compute the Correlation Matrix for the Overall Dataset.
print(corr_matrix)                                                              #Printing out the Correlation Matrix.
```

## 5.1.3 : MOST CORRELATED - LOWEST CORRELATED FEATURES

We display the number of most and lowest correlated features for each possible value of the correlation (threshold). 

```python
#DEFINYING THE FUNCTION FOR THE CORRELATION COMPUTATIONS:
def slide_bar_(corr_matrix, t0):                                        #Definition of the function
    c=0                                                                 #definying the counter to be 0.
    for row in corr_matrix:                                             #For each row in the Correlation Matrix.
        for el in row:                                                  #For each element in the Row.
            if el>t0:                                                   #If the Correlation Value is higher than the setted threshold.
                c+=1                                                    #Add up to the counter 1.
    c = c / 2                                                           #Divide by 2, since we have mirrored features.
    print( f'The number of Higher Correlated Feautures is {c}')         #Printing out the Number of Higher Correlated Features.
    print( f'The number of Lower Correlated Features is {50 - c}')      #Printing out the Number of Lower Correlated Features.


#INSERTING A WIDGET
@widgets.interact( t0 = (  0,  1, .005) )                               #We add a slider that takes value from 0 to 1.
def update(t0 = 1):                                                     #We define an update function that calls our 'slide_bar' function, and we set it initially to 1.
    slide_bar_(corr_matrix,t0)                                          #Update the SlideBar(Correlation Matrix aargument in 'slide bar).
```
