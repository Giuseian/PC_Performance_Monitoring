This is achieved thanks the "ll" function that it is able to return the Log_likelihood Function. We can then apply it to every generic *x* given in input. The second step to do is instead to apply the *log-likelihood function l(a)* directly to the specific *x20*:
$$x_{20} = {[1.7, 5.1, 2.5, 5.6, 6, 0.3, 3.3, 5.7, 3.5, 2.8, 6, 2.8, 4.4, 3.7, 0.6, 5.7, 1.4, 0.2, 1.9, 6]}$$



### 2.5.1: IMPLEMENTING THE *LOG-LIKELIHOOD FUNCTION l(a)*:

```python
#IMPORTING LIBRARIES:
import numpy as np                                  #We use it for dividing the space linearly and for doing some array's computations.

#IMPLEMENTATING THE Log-Likelihood-Function:  
def ll (alpha,x):                                   #Definition of the "ll" function taking in input the alpha and x values.
    sum_ = 0                                        #Initializing a varible "sum" to 0.
    for el in x:                                    #Looping in the x arrays and going through all the values in x.
        sum_+=np.log(1+alpha*(np.cos(el)))          #Summing up the log likelihood of the function.
    return sum_                                     #Returning the "sum" variable.
```

### 2.6.1: FIND THE MAXIMUM LIKELIHOOD ESTIMATE **(MLE)** USING **L-BFGS** STANDARD:

#### TASK:
*What about the maximum likelihood estimator for ${\alpha}$? Well, letâ€™s use **L-BFGS** to find the MLE estimate numerically based on ${X_{20}}$ with and without gradient information (notice that is bounded between âˆ’1/3 and +1/3). Start the optimizer at multiple initial points and then report and comment your results (also in comparison with the method of moments solution).*

#### IMPLEMENTATION APPROACH:
In order to find the *MLE* for $\alpha$ we maximize the function recently found: the *Log-Likelihood*. In order to do this, we will apply the following walk-around: we will minimize the negative log-likelihood. 

We apply the Minimization using a specific method, the **BFGS(Broydenâ€“Fletcherâ€“Goldfarbâ€“Shannon) Algorithm**, namely an iterative method that aims in solving unconstrained nonlinear optimization problems. In particular, we opt for a specific "version" of it that is the **L-BFGS-B** that stands for "Limited-Memory BFGS" in which we accept a solution that is slightly worse in terms of precision but a little bit faster in terms of iterations. Indeed, sometimes we are dealing with large problems and optimization is, in turn, heavy. For this reason, there are less "hungry" memory types of methods, that goes under the BFGS umbrella, that uses less memory footprint, like the L-BFGS.

Finally, we then proceed to plot the *Maximum Log-Likelihood* with the *Maximum Likelihood Estimate* (plotted in red) through the Usage of *MatPlotLib*. 


```python
#IMPORTING LIBRARIES:
import matplotlib.pyplot as plt                         #We use it in order to simply plot our requested Density.  
import numpy as np                                      #We use it for dividing the space linearly and for doing some array's computations.
from scipy.optimize import minimize                     #Specific routine for Optimization. This package provides functions for minimizing (or maximizing) objective functions, possibly 
                                                        #subject to constraints. 
                                                        
#DEFINITION OF SOME VARIABLES:
x = np.array([1.7, 5.1, 2.5, 5.6, 6, 0.3, 3.3, 5.7, 3.5, 2.8, 6, 2.8, 4.4, 3.7, 0.6, 5.7, 1.4, 0.2, 1.9, 6])
r_min, r_max = -1/3, 1/3                                #These are the two bouunds in which we search for the value.
                                                        #In this case we are minimizing(we already know its minimum is at zero). We can just take the "-likelihood" or the "-loglikelihood".

#OBJECTIVE FUNCTION:
def nll(alpha):                                         #Negative Likelihood Function. This constitutes the "Objective Funtion" we want to optimize.
    return -ll(alpha,x)                                 #It returns the negative of the log-likelihood, that we will use in minimization. (Walk-around)

#DIVISION OF THE SPACE:                                 #Defining the Search Space(the box we're going to look for our maximum likelihood estimate). 
alpha = np.linspace(-1/3,1/3,num=20)                    #We save in "x" an evenly spaced sequence of numbers(20 numbers in this case) over a specified interval(In our case [-1/3,1/3]).

#SETTING THE SEED:
np.random.seed(34143)                                   #If you are not setting the seed, that is the initial point for you random number generator(pseudo-random number generators), each 
                                                        #time you will run the code, you are going to get different numbers. This is useful for making Reproducible our code, in such a way 
                                                        #that whoever run the code, gets the same result.

#MINIMIZING:                                                        
pt = r_min + np.random.rand(1) * (r_max - r_min)        #We create the initialize at random an array starting points.
result = minimize(nll, pt, method = 'L-BFGS-B')         #We apply the Minimization to the Negative Log-Likelihood, since out aim is to find the Maximum Likelihood Estimate. In particular,
                                                        #we do this, using the BFGS method.

#PLOTTING:                                              
plt.title("Maximum Likelihood Estimate(L-BFGS-B)")      #Inserting the Title of the plot.
plt.xlabel('alpha')                                     #Inserting the x-axis' label in the plot.
plt.ylabel('log-likelihood')                            #Inserting the y-axis' label in the plot.
plt.plot(alpha,Z)                                       #Plotting the Log-Likelihood Function.
plt.scatter(result['x'],ll(result['x'],x),color='red')  #Adding to it the Maxium Likelihood Estimation for the plot that we have found using the L-BFGS-B.

#PRINTING OF THE MAIN RESULTS OBTAINED:
solution = result['x']                                              #Selecting the precise point in whichh we have the Maximum Likelihood Estimate from the result obtained. 
evaluation = nll(solution)                                          #We save the evaluation of the Negative Log-Likelihood.

#EVALUATION OF THE RESULTS OBTAINED:       
print("The Maximum Likelihood Estimate done with L-BFGS is at: %.8f" % result['x'])  #We print out the MLE result explicitly.
print('Status of the Convergence : %s' % result['message'])                          #We print out the message's status about the algorithm convergence.
print('Total Number of Evaluations: %d' % result['nfev'])                            #We print out the number of iterations that the Optimization algorithm does in total.
print('Solution: f(%s) = %.5f' % (solution, evaluation))                             #We print out the Maximum Likelihood. 
```

### 2.6.2: FIND THE MAXIMUM LIKELIHOOD ESTIMATE **(MLE)** USING **L-BFGS** WITH THE JACOBIAN SPEED-UP:

We will use the derivative(Gradient), or Jacobian of our Objective funtion, with the aim to speed up our methods. The fact that it is a better solution, is due to the fact that when you are evaluating the MLE with the Jacobian argument directly, it is a better solution because here we are evaluating the gradient exactly, whilst if the Jacobian argument is not indicated, we are evaluating the gradient numerically, and thus with way less precision. 

```python
#IMPORTING LIBRARIES:
import numpy as np                                      #We use it for dividing the space linearly and for doing some array's computations.

#DEFINITION OF OUR X_20 ARRAY:
x = np.array([1.7, 5.1, 2.5, 5.6, 6, 0.3, 3.3, 5.7, 3.5, 2.8, 6, 2.8, 4.4, 3.7, 0.6, 5.7, 1.4, 0.2, 1.9, 6])

#DERIVATIVE FUNCTION DEFINITION:
def derivative(alpha):                                  #The derivative(Gradient), or Jacobian of our Objective funtion, will be useful to speed up our methods.                          
    d=0                                                 #Initializing the sum variable that will handle the sum of the derivatives computed in the samples x_i.
    for el in x:                                        #Iterating for every sample point x_i in the x_20 numpy array of samples.
        d+=np.cos(el)/(np.cos(el)*alpha+1)              #Adding to our sum variable the derivative evaluation at point x_i.
    return np.array(-d)                                 #Finally, returning the sum of the derivatives evaluated at the sample points.

#MINIMIZATION:
result = minimize(nll, pt, method = 'L-BFGS-B', jac = derivative)   #Applying the Minimization, as before, to the Negative Log-Likelihood. This time, though, we make use of a more 
                                                                    #powerful tool, the derivatives. Indeed, thanks to the use of the "Jacobian" argument in the Minimize function of 
                                                                    #the Optimization, we now get something way faster than before in terms of evaluations(6 vs 12) and also way closer 
                                                                    #to the solution. In the end, notice that this is a Local Optimization and not a Global one!
                                                                    
#PRINTING OF THE MAIN RESULTS OBTAINED:
solution = result['x']                                              #Selecting the precise point in whichh we have the Maximum Likelihood Estimate from the result obtained. 
evaluation = nll(solution)                                          #We save the evaluation of the Negative Log-Likelihood.         

#EVALUATION OF THE RESULTS OBTAINED:       
print("The Maximum Likelihood Estimate done with L-BFGS and using the Jacobian Speed-up is at: %.8f" % result['x'])  #We print out the MLE result explicitly.
print('Status of the Convergence : %s' % result['message'])          #We print out the message's status about the algorithm convergence.
print('Total Number of Evaluations: %d' % result['nfev'])            #We print out the number of iterations that the Optimization algorithm does in total.
print('Solution: f(%s) = %.5f' % (solution, evaluation))             #We print out the Maximum Likelihood. 
```

### 3.4.3.1: CROSS-VALIDATION ON TWO SEPARATE GAUSSIANS

#### IMPLEMENTATION APPROACH:
In this step, we first apply the **Cross Validation** separately to the two Values number_cyl and city_mpg. 

```python
#IMPORT LIBRARIES:
from scipy.stats import norm                                        #We will use it for computing the Gaussian Pdf. 
import numpy as np                                                  #We use it for some array's computations.
import sklearn                                                      #We will use it for the Gaussian Mixtures.

#GENERATING DIFFERENT GAUSSIAN MIXTURE:
n_components = np.arange(1, 6)                                      #We create an array that defines the number of gaussian mixtures we are going to generate(in this case 5 Gaussians).
models = [sklearn.mixture.GaussianMixture(n_components = n, covariance_type='diag', random_state=0).fit(df_tr) for n in n_components] #We generate n Gaussian Mixtures. 
listofparameters=[]                                                 #We initialize a list to fill with a triple of parameters (weight,mean,covariance) to be taken from each gaussian
                                                                    #mixture.
for model in models:                                                #Looping over the models.
    listofparameters.append(( np.round(model.weights_, decimals=2), np.round(model.means_,decimals=2), np.round(model.covariances_,decimals=2)))    #Append the triple(weight, mean,
                                                                                                                                                    #covariance).

#GAUSSIAN PROBABILITY DENSITY FUNCTION:                                 
def g_pdf(samp,m,sd):                                               #This function returnes the density of a gaussian given: a point, the mean and the standar deviation.
    return norm.pdf(samp,loc=m,scale=sd)                            

#MIXTURE OF GAUSSIANS PROBABILITY DENSITY FUNCTION:
def mg_pdf(samp,gm):                                                #This function returnes the density of a mixture of gaussians given a gaussian mixture.
    s=0                                                             #Initializing a sum equal to zero.
    for i in np.arange(gm.n_components):                            #Looping over the number of components.
        w=gm.weights_[i]                                            #Assign to "w" the weight of the n-th gaussian mixture.
        s+=w*g_pdf(samp,gm.means_[i],np.sqrt(gm.covariances_[i]))   #Assign to "s" the sum of the gaussian densities multiplied for their weights (that is the definition of density of a 
                                                                    #mixture of gaussians)
    return s                                                        
    
#CROSS VALIDATION FUNCTION:
def CV (data_test,gm):                                              #Cross Validation: Calculating cv given a fixed list of gaussians (so a single mixture of gaussians).
    cv=0                                                            #Setting up cv variable to 0.
    values=data_test.values                                         #"values" takes values from the data_test and obtains a list of couples of values, namely: numb_of_cyl and of city_mpg. 
    for x in values:                                                #Looping over the x values in values: 
        cv+=np.log(mg_pdf(x,gm))                                    #We sum in cv the logarithm of the density of the mixture of gaussians.
    cv=cv/len(values)                                               #Once we've finished summing we divide by the number of mixtures.
    return cv                                                       #It rturns a list with two cv: one for the num_of_cyl and the other for city_mpg.
print(models)

#PRINTING CV VALUES FOR DIFFERENT MIXTURES OF GAUSSIANS:
print('CV1 IS', CV(df_te,models[0]))                                #Print out the values for the Mixtures of Gaussian 0.            
print('CV2 IS',CV(df_te,models[1]))                                 #Print out the values for the Mixtures of Gaussian 1.  
print('CV3 IS',CV(df_te,models[2]))                                 #Print out the values for the Mixtures of Gaussian 2.  
print('CV4 IS', CV(df_te,models[3]))                                #Print out the values for the Mixtures of Gaussian 3.  
print('CV5 is', CV(df_te,models[4]))                                #Print out the values for the Mixtures of Gaussian 4.  

#ACCURACY USING SCORE:
label = models[2].predict(df_te)                                    #We save the labels of a specific Gaussian in "label".
a = np.array(df_te)                                                 #We create a numpy array with the test samples. 
acc = models[2].score(a, label)                                     #We obtain an accuracy score applied to a certain model. 
print(acc)                                                          #We print the accuracy obtained from our model.

#PLOTTING
plt.plot(n_components, [m.aic(df1) for m in models], label='AIC')   #Plotting the AIC Method applied to our models.
plt.legend(loc='best')                                              #Inserting the legend to our plot.
plt.xlabel('n_components')                                          #Inserting the label to the x-axis of our plot.
```

### 3.5: PLOTTING THE MIXTURE OF GAUSSIAN WITH THE MoG RESULT:

#### TASK:
*Repeat the analysis using k-mean by selecting k via the Elbow method. Compare with the previous results.*
#### IMPLEMENTATION APPROACH:
In this step we individuate the right number of clusters that will be used in the **K-MEANS ALGORITHM** thanks to the **ELBOW-METHOD**, that is a heuristic used in determining the number of clusters in a data set(showing the optimal k). The method consists of plotting the explained variation as a function of the number of clusters, and picking the elbow of the curve as the number of clusters to use. In our case, the real Elbow occurs at the value "2". However, there is also an additional significant inclination forming another Elbow at "3". We thus opt for the three clusters, since it is the real Elbow in the function, and then augmenting the k, does not imply so much difference in the distortion..

Notice that, indeed, if we had chosen as value 2, we would have had a very great distortion, w.r.t. have chosen 3 clusters as in our case! The more we increase k, the more we restrict the distortion. Notice that the more k increases, the more the distortion flatten, and thus it is not convenient anymore to choose a bigger value if the change in distortion is Minimal. In our case, the great change is between 1 and 2, but also between 2 and 3 another iportant change occurs. Notice that we could have also chosen 2 clusters,  if we accept to have a significant distortion.
#### COMMENT:
Notice that, as the last plot visualizes, this is a Tipical example where the K-Means suffer. Indeed, we do not have "Blobs", but straight lines. We can see, that the clustering of the K-MEANS, in this case, really obtain a real bad clusteirng. 

```python
#IMPORT LIBRARIES
from sklearn.cluster import KMeans                                  #We use it in order to apply the K-Means.

#K-Means and Elbow-Method:
distortions = []                                                    #Initializing the list "distorsions" with all values in in it, that will be the values of the kmeanModel.
K = range(1,6)                                                      #Initializing K as a range from 1 to 6.
for k in K:                                                         #Looping over the ks.
    kmeanModel = KMeans(n_clusters=k)                               #Applying the Kmeans with k number of clusters.
    kmeanModel.fit(df1)                                             #Check how they are with the Dataset, or better, how they fit.
    distortions.append(kmeanModel.inertia_)                         #Append their value in the distortions list, representing the Distortions that we have if we choose a model with 
                                                                    #k clusters.
    
#PLOTTING:
plt.plot(K, distortions, 'bx-')                                     #Plotting the Elbow Methos, with K Clusters and the Distortions.
plt.xlabel('k')                                                     #Inserting the x-axis label.
plt.ylabel('Distortion')                                            #Inserting the y-axis label.
plt.title('The Elbow Method')                                       #Inserting the title in the plot.
plt.show()          

#APPLYING K-MEANS CHOOSING THE K DEPENDING ON THE ELBOW METHOD HEURISTIC: 
kmeans = KMeans(n_clusters=3, random_state=0)                       #We select 3 clusters taking this value from the Elbow Method.
klabs = kmeans.fit_predict(df1)                                     #Saving the Labeling of the K-Means.

plt.scatter(df1. number_cyl, df1.city_mpg, c=klabs, s=40, cmap='viridis')   #Plotting the points in every k values.
```
